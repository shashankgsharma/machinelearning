```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x_train = np.array([ 1, 2 ])
y_train = np.array([ 300, 500 ])
df = pd.DataFrame()
df['Size (1000 sqft)'] = x_train
df["Price (1000's of dollars)"] = y_train
df
```

```python
print(f"x_train: {x_train}\ny_train: {y_train}")
```

## Number of training examples = m

```python
# x_train.shape[0]
# or
print(f"Number of training examples(m) = {len(x_train)}")
```

## Training example x_i, y_i

```python
i = 1
x_i = x_train[i]
y_i = y_train[i]

print(f"x^({i}), y^({i}) = ({x_i}, {y_i})")
```

## Plotting the Data

```python
plt.scatter(x = x_train, y = y_train, c = ['red', 'green'], marker = 'x')
plt.title('House Pricing Details', fontdict = {'fontname': 'arial', 'fontsize': 15, 'color': 'blue', 'fontweight': 'bold'})
plt.xlabel('Size (1000 sqft)')
plt.ylabel("Price (1000's of dollars)")

plt.plot()
```

```python
def compute_model_output(x, w, b):
    
    m = len(x)
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w*x[i] + b
        
    return f_wb
```

```python
w = 200
b = 100

y_hat = compute_model_output(x_train, w, b)    #y_hat = predicted value of output variable y

plt.plot(x_train, y_hat, color = 'red', label = 'Function Prediction')
plt.scatter(x_train, y_train, color = 'green', marker = 'x', label = 'Actual Data')
plt.title('Housing Prices', fontdict = {'fontname': 'Algerian', 'fontsize': 20, 'fontweight': 'bold', 'color': 'Brown'})
plt.xlabel('Size (1000 sqft)')
plt.ylabel('Price (1000 $)')

plt.legend()
plt.show()
```

```python
# Prediction of price of house with 1200 sqft area

x_i = [1.2]
cost_1200sqft = compute_model_output(x_i, w, b)[0]
print(f'Cost of House with 1200sqft area: ${cost_1200sqft:.0f},000')
```
