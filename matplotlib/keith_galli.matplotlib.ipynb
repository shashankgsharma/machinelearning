## Loading necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Line Graph

```python
x1 = [1, 2, 3, 4]
y1 = [2, 4, 6, 8]

#Resize graph
plt.figure(figsize = (5, 3), dpi = 100)    #dpi = pixels per inch --- 1500px x 900px

# plt.plot(x, y, label = '2x', color = 'orange', linewidth = 2, marker = '.', markersize = 10, linestyle = 'dashdot', markeredgecolor = 'blue')

#shorthand notation for the above line: 
#fmt = '[color][maker][linestyle]'
                #or
#      '[marker][linestyle][color]'    ## preferred ##

plt.plot(x1[0:2], y1[0:2], 'o-b', label = '2x')
plt.plot(x1[1:], y1[1:], 'o-.b')

#2nd Line
x2 = np.arange(1, 4, 0.5)
print(x2)
plt.plot(x2[0:3], x2[0:3] ** 2, '^-r', label = 'x^2')
plt.plot(x2[2:], x2[2:] ** 2, '^--r')


plt.title('Line Graph', fontdict = {
                                    'fontname': 'algerian',
                                    'fontsize': 20,
                                     'fontweight': 'bold',
                                     'color': 'red',
                                     'verticalalignment': 'baseline',
#                                      'horizontalalignment': 'baseline'
                                    })
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.xticks(ticks = [1, 2, 3, 4])
# plt.yticks(ticks = [2, 4, 6, 8])
plt.legend()

plt.savefig('linegraph.png', dpi = 300)

plt.show()
```

## Bar Graph

```python
labels = ['A', 'B', 'C', 'D']
values = [5, 1, 0.5, 9]

plt.figure(figsize = (6, 4), dpi = 100)
plt.title('Bar Graph', fontdict = {
                                    'fontname': 'Algerian', 
                                    'fontsize': 20, 
                                    'fontweight': 'bold', 
                                    'color': 'purple'
                                  })
plt.xlabel('Classes')
plt.ylabel('A-graders')
bars = plt.bar(labels, values, color = ['green', 'orange', 'grey', 'red'], edgecolor = 'black')

patterns = ['*', '+', 'O', '/']
for bar in bars: 
    bar.set_hatch(patterns.pop())
    
# bars[0].set_hatch('/')
# bars[1].set_hatch('*')
# bars[2].set_hatch('O')
# bars[3].set_hatch('X')

plt.show()
```

## Load FIFA data

```python
fifa = pd.read_csv('fifa_data.csv')
fifa
```

### Histograms

```python
bins = np.arange(40, 110, 10)

plt.hist(fifa.Overall, bins = bins, color = '#ababf0', align = 'mid')
plt.title('FIFA Player Skill levels 2018')
plt.xlabel('Skill Level')
plt.ylabel('No. of Players')
plt.xticks(bins)

plt.show()
```

### Pie Charts

```python
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
print('Right: ', right)
print('Left: ', left)
```

```python
labels = ['Right Footed Players', 'Left Footed Players']
colors = ['orange', 'green']

plt.title('Foot Preference of FIFA Players', fontdict = {'fontname': 'Calibri', 'fontsize': 25, 'color': 'red', 'fontweight': 'bold'})
plt.pie([right, left], colors = colors, autopct = '%.2f %%')

plt.legend(labels)
plt.show()
```

```python
fifa_weight_kg = []

for weight in fifa.Weight:
    try:
        weight_str = str(weight)
        weight_value = int(weight_str[0:3])
        weight_kg = round(weight_value * 0.45359237)
        fifa_weight_kg.append(weight_kg)
    except (ValueError, TypeError):
        # Handle non-numeric or invalid values
        fifa_weight_kg.append(None)  # or any other appropriate action

fifa_weight_kg
```

```python
fifa.Weight = fifa_weight_kg
fifa.Weight
```

```python
light = len(fifa.loc[fifa.Weight <= 55])
light_medium = len(fifa.loc[(fifa.Weight > 55) & (fifa.Weight <= 65)])
medium = len(fifa.loc[(fifa.Weight > 65) & (fifa.Weight <= 75)])
heavy_medium = len(fifa.loc[(fifa.Weight > 75) & (fifa.Weight <= 85)])
heavy = len(fifa.loc[fifa.Weight > 85])
print('light: ', light)
print('normal: ', light_medium)
print('medium: ', medium)
print('heavy_medium: ', heavy_medium)
print('heavy: ', heavy)
```

```python
weight_categories = [light, light_medium, medium, heavy_medium, heavy]
labels = ['light', 'light medium', 'medium', 'heavy_medium', 'heavy']
colors = ['orange', 'pink', 'gray', 'green', 'yellow'] 
explode = (0.4, 0.4, 0, 0, 0.4)

plt.title('Weight Distribution of FIFA Players (in kg)', fontdict = {'fontname': 'Arial', 'fontsize': 20, 'color': 'red', 'fontweight': 'bold'})
plt.pie(weight_categories, labels = labels, colors = colors, autopct = '%.2f %%', explode = explode)
# plt.legend(labels, loc = 'lower left')

plt.savefig('weight_fifa.png', dpi = 300)
plt.show()
```

## Comparing differnt teams with box and whiskers chart

```python
fifa
```

```python
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
man_united = fifa.loc[fifa.Club == 'Manchester United']['Overall']
man_city = fifa.loc[fifa.Club == 'Manchester City']['Overall']
labels = ['Barcelona', 'Man United', 'Man City']
teams = [barcelona, man_united, man_city]

plt.figure(figsize = (5, 7))
plt.title('FIFA Team Comparisons', fontdict = {'fontname': 'Calibri', 'fontsize': 25, 'color': 'red', 'fontweight': 'bold'})
boxes = plt.boxplot(teams, labels = labels, patch_artist = True, medianprops = {'linewidth': 2, 'color': 'black'})
plt.ylabel('FIFA Overall Raitng')

colors = ['blue', 'brown', 'red']
for box in boxes['boxes']:
    #Set edge color
    box.set(color = colors.pop(0), linewidth = 2)
    
    #Change fill color
    box.set(facecolor = '#cdd1d1')

plt.savefig('fifa_team_compare.png', dpi = 300)
plt.show()
```
